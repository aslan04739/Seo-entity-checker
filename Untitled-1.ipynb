```python
from sentence_transformers import SentenceTransformer

# 1. Load the Model
# "trust_remote_code=True" is required for Stella because it uses a custom architecture.
print("Loading Stella... (this will take a minute the first time)")
model = SentenceTransformer("dunzhang/stella_en_400M_v5", trust_remote_code=True)

# 2. Define your SEO Titles
titles = [
    "Best Running Shoes for Flat Feet 2025",  # A
    "Top Rated Sneakers for Flat Arches",     # B (Semantically identical to A)
    "How to Bake a Chocolate Cake"            # C (Totally different)
]

# 3. Create Embeddings (Turn text into numbers)
# Stella needs a prompt to know it's comparing similarity.
prompt = "Instruct: Retrieve semantically similar text.\nQuery: "
embeddings = model.encode(titles, prompt_name="s2s_query")

# 4. Compare A to B, and A to C
similarity_A_B = model.similarity(embeddings[0], embeddings[1])
similarity_A_C = model.similarity(embeddings[0], embeddings[2])

print(f"\nSimilarity (Shoes vs Sneakers): {similarity_A_B.item():.4f}")
print(f"Similarity (Shoes vs Cake):     {similarity_A_C.item():.4f}")
```
